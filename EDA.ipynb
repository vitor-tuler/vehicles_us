```python
import pandas as pd
import plotly_express as px
```

```python
df = pd.read_csv('./vehicles.csv')
df.head()
```

```python
print(df.info())
```

```python
df.sample(n=10)
```

```python
print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
null_columns = df.columns[df.isna().any()]
```

```python
df.fillna({'is_4wd': 0}, inplace=True)
```

```python
median_year = df['model_year'].median()

df.fillna({'model_year': median_year}, inplace=True)
print(median_year)
```

```python
mediana_hodometro = df['odometer'].median()
df.fillna({'odometer':mediana_hodometro}, inplace=True)
```

```python
mediana_cilindros = df['cylinders'].median()
df.fillna({'cylinders': mediana_cilindros}, inplace=True)
df['cylinders'] = df['cylinders'].astype(int)
```

```python
for i in df.columns:
    print(f"{i} : {df[i].unique().tolist()}")
```

```python
df.fillna({'paint_color': 'Unknown'}, inplace=True)
```

```python
print(df.info())
```

```python
fig_hist = px.histogram(df, x="odometer")
fig_hist.show()
```

O alto volume de valores ausentes, que foram imputados pela mediana, gerou um pico dentro do grfico

```python
fig_scatter = px.scatter(df, x="odometer", y="price")
fig_scatter.show()
```
